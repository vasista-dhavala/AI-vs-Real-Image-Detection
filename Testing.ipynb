```python
# -----------------------------
# Step 0: Load the saved model
# -----------------------------
from tensorflow.keras.models import load_model
import os
import numpy as np
from PIL import Image

from google.colab import drive
drive.mount('/content/drive')

# Path to your saved model
model_path = "/content/drive/MyDrive/image_identifier_model.h5"
model = load_model(model_path)
print("✅ Model loaded successfully!")

# -----------------------------
# Step 1: Preprocess function
# -----------------------------
IMG_SIZE = (224, 224)

def preprocess_image(image_path):
    img = Image.open(image_path).convert('RGB')
    img = img.resize(IMG_SIZE)
    arr = np.array(img) / 255.0
    return arr

# -----------------------------
# Step 2: Prediction with confidence
# -----------------------------
def predict_image_confidence(image_path):
    """Predict if image is AI or Real with confidence."""
    img = preprocess_image(image_path)
    img = np.expand_dims(img, axis=0)
    pred_prob = model.predict(img)[0][0]
    if pred_prob > 0.5:
        label = "Real Image"
        confidence = pred_prob * 100
    else:
        label = "AI-Generated Image"
        confidence = (1 - pred_prob) * 100
    return label, confidence

# -----------------------------
# Step 3: CLI in Colab
# -----------------------------
print("=== AI vs Real Image Identifier ===")
print("Enter 'exit' to quit.\n")

while True:
    image_path = input("Enter image path: ").strip()
    if image_path.lower() == "exit":
        print("Exiting...")
        break
    if not os.path.exists(image_path):
        print("❌ File not found. Try again.\n")
        continue
    label, confidence = predict_image_confidence(image_path)
    print(f"✅ Prediction: {label} ({confidence:.2f}% confident)\n")
```
